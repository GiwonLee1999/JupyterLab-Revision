iloc: Use index to retrieve a row

```python
df.iloc[[0,1]] # returns from 0 to 1
df.iloc[[0,1], 2] # returns from 0 to 1 and second column elements only
```

loc: Returns Columns and rows elements

```python
df.loc[0] # First row with the column names
df.loc[[0,1]] # First to second row
df.loc[[0,1], 'Age'] # First to second row/ Age columns
df.loc[[1], ['Age', 'Employment']] # First to second row/ Age columns
```
